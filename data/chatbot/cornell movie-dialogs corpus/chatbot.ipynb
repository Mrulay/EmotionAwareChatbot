```python
lines_to_preprocess = [x[-1] for x in lines]
lineIDs =  [x[0] for x in lines]
processed_lines = []
for line in lines_to_preprocess:
    processed_lines.append(preProcessText(line))
processed_lines
```

```python
lineDict = dict(zip(lineIDs,processed_lines))
lineDict
```

```python
for conv in convs:
    for i,id in enumerate(conv):
        conv[i] = lineDict[id]
convs
        
```

```python
ques, ans = [], []
for conv in convs:
    if len(conv)%2 != 0:
        conv.remove(conv[-1])

for conv in convs:
    for i in range (0, len(conv)):
        if i%2 == 0:
            ques.append(conv[i])
        elif i%2 != 0:
            ans.append(conv[i]) 
            
```

```python
ques = ques[:40000]
ans = ans[:40000]
```

```python
wordcount = dict()

for line in ques:
    for word in line.split():
        if word not in wordcount:
            wordcount[word] = 1
        else: 
            wordcount[word] += 1

for line in ans:
    for word in line.split():
        if word not in wordcount:
            wordcount[word] = 1
        else: 
            wordcount[word] += 1
```

```python
t = 5
vocab = {}
word_num = 0
for word, count in wordcount.items():
    if count >= t:
        vocab[word] = word_num
        word_num += 1
vocab
```

```python
for i in range(len(ans)):
    ans[i] = '<SOS> ' + ans[i] + ' <EOS>'

tokens = ['<PAD>', '<EOS>', '<OUT>', '<SOS>']
x = len(vocab)
for token in tokens:
    vocab[token] = x
    x += 1
```

```python
ans
```

```python
vocab['cameron'] = vocab['<PAD>']
vocab['<PAD>'] = 0

inv_vocab = {w:v for v, w in vocab.items()}
```

```python
ids = []
for i,line in enumerate(ques):
    x = len(line.split())
    if x > 15:
        ques.remove(line)
        ids.append(i)
for id in ids: 
    del ans[id]
```

```python
tfidf = TfidfVectorizer()
response = tfidf.fit(processed_lines)
idf = tfidf.idf_
tfidf_values = (dict(zip(tfidf.get_feature_names_out(), idf)))
print (len(tfidf_values))
```

```python
encoder_inp = []
for line in ques:
    lst = []
    for word in line.split():
        if word in vocab and word in tfidf_values:
            lst.append(tfidf_values[word])
        elif word in tokens:
            lst.append(0)
        else:
            lst.append(1)
    encoder_inp.append(lst)
```

```python
decoder_inp = []
for line in ans:
    lst = []
    for word in line.split():
        if word in vocab and word in tfidf_values:
            lst.append(tfidf_values[word])
        elif word in tokens:
            lst.append(0)
        else:
            lst.append(1)
    decoder_inp.append(lst)
```

```python
encoder_inp = pad_sequences(encoder_inp, 15, padding='post', truncating='post')
decoder_inp = pad_sequences(decoder_inp, 15, padding='post', truncating='post')
```

```python
decoder_final_output = []
for i in decoder_inp:
    decoder_final_output.append(i[1:]) 

decoder_final_output = pad_sequences(decoder_final_output, 15, padding='post', truncating='post')
```

```python
decoder_final_output = to_categorical(decoder_final_output, len(vocab))
```

```python
decoder_final_output.shape
```

```python
enc_inp = Input(shape=(15, ))
dec_inp = Input(shape=(15, ))
```

```python
VOCAB_SIZE = len(vocab)
embed = Embedding(VOCAB_SIZE+1, output_dim=50, 
                  input_length=15,
                  trainable=True                  
                  )
```

```python
enc_embed = embed(enc_inp)
enc_lstm = LSTM(400, return_sequences=True, return_state=True)
enc_op, h, c = enc_lstm(enc_embed)
enc_states = [h, c]



dec_embed = embed(dec_inp)
dec_lstm = LSTM(400, return_sequences=True, return_state=True)
dec_op, _, _ = dec_lstm(dec_embed, initial_state=enc_states)

dense = Dense(VOCAB_SIZE, activation='softmax')

dense_op = dense(dec_op)

model = Model([enc_inp, dec_inp], dense_op)
```

```python
model.compile(loss='categorical_crossentropy',metrics=['acc'],optimizer='adam')

model.fit([encoder_inp, decoder_inp],decoder_final_output,epochs=15)
```
